### District Summary

```python
import pandas as pd

#Importing and joining the CSVs
df1 = pd.read_csv("schools_complete.csv")
df2 = pd.read_csv("students_complete.csv")
df3 = pd.merge(df1, df2, how = "inner", on = "school_name" ) 

#Calculating the amount of schools 
total_schools = df1["school_name"].nunique()

#Calculating the amount of students 
total_students = len(df2)

#Calculating the total budget of all the schools 
total_budget =  df1["budget"].sum()

#Calculating the average Math Score of all schools 
avg_math = round(df2["math_score"].mean(),2)

#Calculating the average Reading Score of all schools 
avg_read = round(df2["reading_score"].mean(),2)

#Calculating the passing percentage of Math
total = len(df2) 
math1 = df2["math_score"]
M_pass = math1[math1 >= 70]
Passed = M_pass.tolist()
M_Passed = len(Passed)
MP_Perc = (M_Passed / total)

#Calculating the passing percentage of Reading
total = len(df2) 
read1 = df2["reading_score"]
R_pass = read1[read1 >= 70]
RPassed = R_pass.tolist()
R_Passed = len(RPassed)
RP_Perc = (R_Passed / total) 

#Calculating the passing percentage of both Math & Reading
Both = M_pass & R_pass 
Passed_Both = Both.tolist()
sum = len(df2) * 2

PPerc_Both = (len(Passed_Both) / sum)

#Printing the results in a DataFrame 
District_Summary = pd.DataFrame({

                    'Total Schools': [total_schools],
                    'Total Students': [total_students],
                    'Total Budget': [total_budget],
                    'Average Math Score': [avg_math],
                    'Average Reading Score': [avg_read],
                    '% Pass Math': [round(MP_Perc,2)],
                    '% Pass Reading': [round(RP_Perc,2)],
                    '% Pass Math & Reading': [round(PPerc_Both,2)]
})
District_Summary["Total Students"] = District_Summary["Total Students"].map("{:,}".format)
District_Summary["Total Budget"] = District_Summary["Total Budget"].map("${:,.2f}".format)
District_Summary["% Pass Math"] = District_Summary["% Pass Math"].map('{:.2%}'.format)
District_Summary["% Pass Reading"] = District_Summary["% Pass Reading"].map('{:.2%}'.format)
District_Summary["% Pass Math & Reading"] = District_Summary["% Pass Math & Reading"].map('{:.2%}'.format)

District_Summary
```

### School Summary

```python
import pandas as pd

#Importing the CSVs and joining the two CSVs 
df1 = pd.read_csv("schools_complete.csv")
df2 = pd.read_csv("students_complete.csv")
df3 = pd.merge(df1, df2, how = "inner", on = "school_name" ) 

#creating the empty DataFrame 
School_Summary = pd.DataFrame(columns = [] )

#Print the School Names
nc1 = df1["school_name"].sort_values()
School_Summary["School Name"] = pd.concat([School_Summary, nc1], ignore_index = True)

#Print the School Types and renanming the column
nc2 = df3.groupby(["school_name"])["type"].first().reset_index()
School_Summary = pd.concat([School_Summary, nc2], axis = 1)
School_Summary.rename(columns= {"type" : "School Type"}, inplace = True)
School_Summary.drop("school_name", axis = 1, inplace=True)

#Print the Total Students for each School and renanming the column
nc3 = df3.groupby(["school_name"])["size"].first().reset_index()
School_Summary = pd.concat([School_Summary, nc3], axis = 1)
School_Summary.rename(columns= {"size" : "Total Students"}, inplace = True)
School_Summary.drop("school_name", axis = 1, inplace=True)

#Print the Total Budget for each School and renanming the column
nc4 = df3.groupby(["school_name"])["budget"].first().reset_index()
School_Summary = pd.concat([School_Summary, nc4], axis = 1)
School_Summary.rename(columns= {"budget" : "Total School Budget"}, inplace = True)

#Print the Price per Student for each School and renanming the column
nc5 = School_Summary['Total School Budget'] / School_Summary['Total Students']  
School_Summary = pd.concat([School_Summary, nc5], axis = 1)
School_Summary.rename(columns= {0 : "Per Student Budget"}, inplace = True)

#Print the average Math score for each school
a_math = df3.groupby(["school_name"])["math_score"].mean().reset_index()
a_math.drop("school_name", axis = 1, inplace=True)

School_Summary = pd.concat([School_Summary, round(a_math,2)], axis = 1)
School_Summary.rename(columns= {"math_score" : "Math Score"}, inplace = True)

#Print the average Reading score for each school
a_read = df3.groupby(["school_name"])["reading_score"].mean().reset_index()
a_read.drop("school_name", axis = 1, inplace=True)

School_Summary = pd.concat([School_Summary, round(a_read,2)], axis = 1)
School_Summary.rename(columns= {"reading_score" : "Reading Score"}, inplace = True)

#Counting the number of students for each school
student_count_by_school = df3.groupby(["school_name"])["size"].size()

#Calculating the Math Passing percentage of each school
PP_Math = (((df3[df3["math_score"] >= 70].groupby(["school_name"])["math_score"].count()) / student_count_by_school) * 100).reset_index()
PP_Math.drop("school_name", axis = 1, inplace=True)

nc6 = PP_Math
School_Summary = pd.concat([School_Summary, round(PP_Math,2)], axis = 1)
School_Summary.rename(columns= {0 : "% Passing Math"}, inplace = True)

#Calculating the Reading Passing percentage of each school
PP_Read = (((df3[df3["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()) / student_count_by_school) * 100).reset_index()
PP_Read.drop("school_name", axis = 1, inplace=True)

nc7 = PP_Read
School_Summary = pd.concat([School_Summary, round(PP_Read,2)], axis = 1)
School_Summary.rename(columns= {0 : "% Passing Reading"}, inplace = True)

#Calculating the Math & Reading Passing percentage of each school 
PP_Both = (df3[(df3["math_score"] >=70) & (df3["reading_score"] >=70)].groupby(["school_name"])["school_name"].count() / student_count_by_school * 100).reset_index()
nc8 = PP_Both
School_Summary = pd.concat([School_Summary, round(PP_Both,2)], axis = 1)
School_Summary.rename(columns= {0 : "% Passing Math & Reading"}, inplace = True)
School_Summary.drop("school_name", axis = 1, inplace=True)

School_Summary.set_index('School Name', inplace=True)

#Printing the DataFrame
School_Summary
```
